Определите с помощью стандартных модулей Python средние траты в ресторане в Кировской области за 2017 год.

Выполните обработку ошибок в программе.

Постройте график изменений средних трат в ресторане в Кировской области за 2017.

```python
# С библиотекой pandas
df = pd.read_csv('/Users/a1/Desktop/opendata.csv')
```

```python
df['date'] = pd.to_datetime(df['date']) 
```

```python
# Сортировка df по критериям
df1 = df[(df['name'] == 'Средние траты в ресторане') & (df['region'] == 'Кировская область') & (df['date'].dt.year == 2017)]
```

```python
# Нахождение среднего за 2017 год
df1['value'].mean()
```

```python
df1
```

```python
# Построение графика

# 1.1 Размер графика
from pylab import rcParams

rcParams["figure.figsize"] = 9, 9

# 1.2 Словари для стилей
label_font = {
    "fontsize": 12,
    "family": "serif",
}

title_font = {
    "fontsize": 14,
    "fontweight": "bold",
    "color": "#808080",
    "family": "serif",
}

# 1.3 Сам график
plt.plot(df1['date'].dt.month, df1['value'])
plt.title("График изменений средних трат в ресторане в Кировской области за 2017", fontdict=title_font)
plt.xlabel("Номер месяца", fontdict=label_font)
plt.ylabel("Средние траты в ресторане", fontdict=label_font)

plt.show()
```
